# Feature Engineering

```python
cleaned_df["SalesLineTotal"] = cleaned_df["Quantity"] * cleaned_df["Price"]
cleaned_df
```

```python
aggregated_df = cleaned_df.groupby(by="Customer ID", as_index=False) \
    .agg(
        MonetaryValue=("SalesLineTotal", "sum"),
        Frequency=("Invoice", "nunique"),
        LastInvoiceDate=("InvoiceDate", "max")
    )

aggregated_df.head(5)
```

```python
max_invoice_date = aggregated_df["LastInvoiceDate"].max()

aggregated_df["Recency"] = (max_invoice_date - aggregated_df["LastInvoiceDate"]).dt.days

aggregated_df.head(5)
```

```python
M_Q1 = aggregated_df["MonetaryValue"].quantile(0.25)
M_Q3 = aggregated_df["MonetaryValue"].quantile(0.75)
M_IQR = M_Q3 - M_Q1

monetary_outliers_df = aggregated_df[(aggregated_df["MonetaryValue"] > (M_Q3 + 1.5 * M_IQR)) | (aggregated_df["MonetaryValue"] < (M_Q1 - 1.5 * M_IQR))].copy()

monetary_outliers_df.describe()
```

```python
F_Q1 = aggregated_df["Frequency"].quantile(0.25)
F_Q3 = aggregated_df["Frequency"].quantile(0.75)
F_IQR = F_Q3 - F_Q1

Freqency_outliers_df = aggregated_df[(aggregated_df["Frequency"] > (F_Q3 + 1.5 * F_IQR)) | (aggregated_df["Frequency"] < (F_Q1 - 1.5 * F_IQR))].copy()
Freqency_outliers_df.describe()
```

```python
non_outlier_df = aggregated_df[(~aggregated_df.index.isin(monetary_outliers_df.index)) & (~aggregated_df.index.isin(Freqency_outliers_df.index))]
non_outlier_df.describe()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(data=non_outlier_df['MonetaryValue'], color='skyblue')
plt.title('Monetary Value Boxplot')
plt.xlabel('Monetary Value')

plt.subplot(1, 3, 2)
sns.boxplot(data=non_outlier_df['Frequency'], color='lightgreen')
plt.title('Frequency Boxplot')
plt.xlabel('Frequency')

plt.subplot(1, 3, 3)
sns.boxplot(data=non_outlier_df['Recency'], color='salmon')
plt.title('Recency Boxplot')
plt.xlabel('Recency')

plt.tight_layout()
plt.show()
```

# KMeans Clustering

```python
max_k = 12
inertia = [] 
silhouette_scores = []
k_values = range(2, max_k + 1) 

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, max_iter=1000)
    
    cluster_labels = kmeans.fit_predict(scaled_data_df)
    sil_score = silhouette_score(scaled_data_df,cluster_labels)
    silhouette_scores.append(sil_score)
    inertia.append(kmeans.inertia_)

# Plotting should happen after the loop
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.plot(k_values, inertia, marker='o')
plt.title('KMeans Inertia for different values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)


plt.subplot(1,2,2)
plt.plot(k_values, silhouette_scores, marker='o',color = 'orange')
plt.title('Silhouetter Scores for different values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhhouetter Score')
plt.xticks(k_values)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=4,random_state=42,max_iter=1000) 
cluster_labels = kmeans.fit_predict(scaled_data_df) 
cluster_labels
```

```python
non_outlier_df = non_outlier_df.copy()
non_outlier_df['Cluster'] = cluster_labels
non_outlier_df
```

```python
cluster_colors = {
    3: '#ff3333', #red
    1: '#ffaf33', #orange 
    2: '#4fff33', #green 
    0: '#3374ff', #blue
}

colors = non_outlier_df['Cluster'].map(cluster_colors) 
fig = plt.figure(figsize=(10,10)) 
ax = fig.add_subplot(projection = '3d') 
scatter = ax.scatter(non_outlier_df['MonetaryValue'],
                     non_outlier_df['Frequency'],
                     non_outlier_df['Recency'],
                     c = colors, marker = 'o') 

ax.set_xlabel('MonetaryValue') 
ax.set_ylabel('Frequency') 
ax.set_zlabel('Recency') 

ax.set_title('3d Scatter Plot of Customers Data by Cluster') 
plt.show()
```

```python
plt.figure(figsize=(12,18)) 

plt.subplot(3,1,1) 
sns.violinplot(x=non_outlier_df['Cluster'], y= non_outlier_df['MonetaryValue'], palette=cluster_colors,hue=non_outlier_df['Cluster'])
sns.violinplot(y=non_outlier_df['MonetaryValue'],color='gray',linewidth=1.0) 
plt.title = ('MonetaryValue by Cluster') 
plt.ylabel = ('MonetaryValue') 

plt.subplot(3,1,2) 
sns.violinplot(x=non_outlier_df['Cluster'], y= non_outlier_df['Frequency'], palette=cluster_colors,hue=non_outlier_df['Cluster'])
sns.violinplot(y=non_outlier_df['Frequency'],color='gray',linewidth=1.0) 
plt.title = ('Frequency by Cluster') 
plt.ylabel = ('Frequency') 

plt.subplot(3,1,3) 
sns.violinplot(x=non_outlier_df['Cluster'], y= non_outlier_df['Recency'], palette=cluster_colors,hue=non_outlier_df['Cluster'])
sns.violinplot(y=non_outlier_df['Recency'],color='gray',linewidth=1.0) 
plt.title = ('Recency by Cluster') 
plt.ylabel = ('Recency') 

plt.tight_layout()
plt.show()
```

# Interpretation 

#### Naming the clusters 

Cluster 0 (Blue)   : These customers make high-value purchases with decent frequency but do not purchase as frequently as the top segment. Their 
                     purchase behavior might be skewed due to irregular buying patterns, but they remain valuable.      
                     Loyal Occasional Shoppers

Cluster 1 (Yellow) : This group has lower engagement, with low monetary value, low frequency, and a significant gap between purchases. They might 
                     require targeted re-engagement campaigns to increase their activity. Infrequent Low Spenders

Cluster 2 (Green)  : Customers in this cluster have recently made purchases but display low frequency and monetary value, similar to Cluster 1. 
                     They may have potential for growth if nurtured properly.
                     Recent Low Spenders

Cluster 3 (Red)    : These are the best customers, purchasing frequently, spending more, and making recent purchases. They are the ideal customers 
                     to retain and continue engaging with. 
                     Top VIP Customers

```python
overlap_indices = monetary_outliers_df.index.intersection(Freqency_outliers_df.index) 

monetary_only_outliers = monetary_outliers_df.drop(overlap_indices)
frequency_only_outliers = Freqency_outliers_df.drop(overlap_indices) 
mon_and_freq_outliers = monetary_outliers_df.loc[overlap_indices] 

monetary_only_outliers['Cluster'] = -1 
frequency_only_outliers['Cluster'] = -2 
mon_and_freq_outliers['Cluster'] = -3 

outliers_clusters = pd.concat([monetary_only_outliers,frequency_only_outliers,mon_and_freq_outliers]) 
outliers_clusters
```

```python
plt.figure(figsize=(12,18)) 

colors_out = {-1:'#FF00FF',
              -2:'#9966FF',
              -3:'#FF0033'}
plt.subplot(3,1,1) 
sns.violinplot(x=outliers_clusters['Cluster'], y= outliers_clusters['MonetaryValue'], palette=colors_out,hue=outliers_clusters['Cluster'])
sns.violinplot(y=outliers_clusters['MonetaryValue'],color='gray',linewidth=1.0) 
plt.title = ('MonetaryValue by Cluster') 
plt.ylabel = ('MonetaryValue') 

plt.subplot(3,1,2) 
sns.violinplot(x=outliers_clusters['Cluster'], y= outliers_clusters['Frequency'], palette=colors_out,hue=outliers_clusters['Cluster'])
sns.violinplot(y=outliers_clusters['Frequency'],color='gray',linewidth=1.0) 
plt.title = ('Frequency by Cluster') 
plt.ylabel = ('Frequency') 

plt.subplot(3,1,3) 
sns.violinplot(x=outliers_clusters['Cluster'], y= outliers_clusters['Recency'], palette=colors_out,hue=outliers_clusters['Cluster'])
sns.violinplot(y=outliers_clusters['Recency'],color='gray',linewidth=1.0) 
plt.title = ('Recency by Cluster') 
plt.ylabel = ('Recency') 

plt.tight_layout()
plt.show()
```

Cluster -3: "Big-Spend Newcomers"
These customers have recently made purchases with high monetary value, but they buy infrequently. They are valuable for their big purchases, though they don't shop often.

Cluster -2: "Dormant Low Spenders"
This group consists of customers who rarely make purchases, with very low spending and little recent activity. They are the least engaged and contribute minimally to revenue.

Cluster -1: "Moderate Spenders with Less Recent Activity"
These customers spend more than those in Cluster -2 but much less than those in higher clusters. While they make purchases more frequently, their spending is still moderate, and their activity is less recent than the other low-spending groups.

```python
C_labels = { 
    0:'Loyal Occasional Shoppers',
    1:'Infrequent Low Spenders', 
    2:'Recent Low Spenders',
    3:'Top VIP Customers',
    -3:'Big-Spend Newcomers',
    -2:'Dormant Low Spenders',
    -1:'Moderate Spenders with Less Recent Activity'}
    

All_Cluster_df = pd.concat([non_outlier_df,outliers_clusters])
All_Cluster_df
```

```python
All_Cluster_df['C_labels'] = All_Cluster_df['Cluster'].map(C_labels) 
All_Cluster_df
```

```python
C_counts = All_Cluster_df['C_labels'].value_counts() 
feature_means = All_Cluster_df.groupby('C_labels')[['Recency','Frequency','MonetaryValue']].mean()

fig, ax1 = plt.subplots(figsize=(12, 8))

# Bar plot for the cluster distribution
sns.barplot(x=C_counts.index, y=C_counts.values, ax=ax1, palette='rocket', hue=C_counts.index)
ax1.set_ylabel('Number of Customers', color='b')
ax1.set_title('Cluster Distribution with Average Feature Values')

# Set the x-ticks and rotate the labels
ax1.set_xticks(range(len(C_counts.index)))  # Set the tick positions
ax1.set_xticklabels(C_counts.index, rotation=90)  # Set the tick labels and rotate

# Create a secondary y-axis for the line plot
ax2 = ax1.twinx()

# Line plot for average feature values
sns.lineplot(data=feature_means, ax=ax2, palette='Set2', marker='o')
ax2.set_ylabel('Average Value', color='g')

plt.show()
```
